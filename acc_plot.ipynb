```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data for two methods
tagging_efficiency = np.linspace(0.5, 1, 100)
misidentification_rate_A = np.exp(-3 * tagging_efficiency)  # Exponential decay for Method A
misidentification_rate_B = np.exp(-4 * tagging_efficiency)  # Faster decay for Method B

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(tagging_efficiency, misidentification_rate_A, label="Method A", color="blue", linestyle="--", marker="o")
plt.plot(tagging_efficiency, misidentification_rate_B, label="Method B", color="green", linestyle="-", marker="x")

# Labels and title
plt.xlabel('Tagging Efficiency (b-jet)', fontsize=12)
plt.ylabel('1 - Rejection (misidentification rate for light jets)', fontsize=12)
plt.title('Tagging Efficiency vs Misidentification Rate', fontsize=14)
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
